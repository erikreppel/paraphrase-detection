```python
import numpy as np
```

```python
train1 = np.load('./small_data/train_sentence1_vecs.npy')
train2 = np.load('./small_data/train_sentence2_vecs.npy')

test1 = np.load('./small_data/test_sentence1_vecs.npy')
test2 = np.load('./small_data/test_sentence2_vecs.npy')

train_labels = np.load('./small_data/train_labels.npy')
test_labels = np.load('./small_data/test_labels.npy')

print len(train1) == len(train2)
print len(test1) == len(test2)
print len(test1[0]) == len(test2[0])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(np.array([1,2,3]), np.array([1,2.4,4]))
```

```python
def compute_cosine(v1, v2):
    distances = []
    for xi, xj in zip(v1, v2):
        distances.append(cosine_similarity(xi, xj))
    distances = map(lambda x: x[0], distances)
    dist_vec = np.array(distances)
    return dist_vec
```

```python
X_train = compute_cosine(train1, train2)
X_test = compute_cosine(test1, test2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(n_jobs=-1)
```

```python
clf.fit(X_train, train_labels)
```

```python
clf.score(X_test, test_labels)
```
