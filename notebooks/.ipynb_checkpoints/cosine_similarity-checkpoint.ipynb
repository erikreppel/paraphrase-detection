```python
import numpy as np
```

```python
train1 = np.load('./small_data/train_sentence1_vecs_eq.npy')
train2 = np.load('./small_data/train_sentence2_vecs_eq.npy')

test1 = np.load('./small_data/test_sentence1_vecs_eq.npy')
test2 = np.load('./small_data/test_sentence2_vecs_eq.npy')

train_labels = np.load('./small_data/train_labels.npy')
test_labels = np.load('./small_data/test_labels.npy')

print len(train1) == len(train2)
print len(test1) == len(test2)
print len(test1[0]) == len(test2[0])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(np.array([1,2,3]), np.array([1,2.4,4]))
```

```python
def compute_cosine(v1, v2):
    distances = []
    for xi, xj in zip(v1, v2):
        distances.append(cosine_similarity(xi, xj))
    distances = map(lambda x: x[0], distances)
    dist_vec = np.array(distances)
    return dist_vec
```

```python
X_train = compute_cosine(train1, train2)
X_test = compute_cosine(test1, test2)
```

```python
ones = filter(lambda x: x == 1, test_labels)
```

```python
# sanity check to make sure we're not just choosing one class
len(ones) / float(len(test_labels))
```

```python
Xt = np.column_stack((X_train, train1, train2))
```

```python
Xtest = np.column_stack((X_test, test1, test2))
```

```python
print train1.shape, train2.shape, X_train.shape
print test1.shape, test2.shape, X_test.shape
print Xt.shape
print train_labels.shape
print Xtest.shape, test_labels.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# clf = SVC(verbose=True)
# clf = LogisticRegression(verbose=1, n_jobs=-1)
clf = GradientBoostingClassifier(verbose=True, n_estimators=1000)

clf.fit(Xt, train_labels)

clf.score(Xtest, test_labels)
```
